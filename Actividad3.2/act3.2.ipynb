```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Bologna_Italia.csv')
df.head()
```

```python
df.info()
```

```python
def valores_nulos(df):
    valores_nulos_cols = df.isnull().sum()
    valores_nulos_df = df.isnull().sum().sum()
    
    return("Valores nulos por columna", valores_nulos_cols,
            "Valores nulos por dataframe", valores_nulos_df)
```

```python
valores_nulos(df)
```

```python
def sustitución_nulos(df):
    float_con_nulos = df.select_dtypes(include=['float64','float'])
    float = float_con_nulos.fillna(round(float_con_nulos.median(), 1))

    int_con_nulos = df.select_dtypes(include=['int64','int'])
    int_sin_nulos = int_con_nulos.fillna(round(int_con_nulos.mean(),1))

    object_con_nulos = df.select_dtypes(include=['object'])
    object_sin_nulos = object_con_nulos.fillna(method="ffill")
    object_sin_nulos = object_sin_nulos.fillna(method='bfill')


    #el dataframe no incluye “datetime y category”

    Datos_sin_nulos = pd.concat([float,int_sin_nulos,object_sin_nulos], axis=1)
    
    return(Datos_sin_nulos)
```

```python
df_sin_nulos=sustitución_nulos(df)
valores_nulos(df_sin_nulos)
```
